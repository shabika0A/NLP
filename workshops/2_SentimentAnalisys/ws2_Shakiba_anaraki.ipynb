```python
# shakiba anaraki 99442047
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from nltk.corpus import movie_reviews
from tqdm import tqdm
import nltk
import spacy
```

```python
nltk.download('movie_reviews')

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Load IMDB Movie Reviews dataset from NLTK
movie_reviews_data = [(movie_reviews.raw(fileid), category) for category in movie_reviews.categories() for fileid in movie_reviews.fileids(category)]

# Extract texts and labels
texts, labels = zip(*movie_reviews_data)

# Convert labels to binary (0 for negative, 1 for positive)
labels = [1 if label == 'pos' else 0 for label in labels]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
# Function to calculate document embeddings using spaCy
def calculate_embeddings(texts):
    embeddings = []
    for text in tqdm(texts):# for aproximate time this function takes
        doc = nlp(text)
        # Average word vectors to get document vector
        # if token.has_vector
        doc_vector = np.mean([token.vector for token in doc ], axis=0)
        embeddings.append(doc_vector)
    return np.array(embeddings)

# Calculate embeddings for training and testing sets
X_train_embeddings = calculate_embeddings(X_train)
X_test_embeddings = calculate_embeddings(X_test)
```

```python
# from sklearn.decomposition import TruncatedSVD
# from sklearn.svm import SVC
# from sklearn.pipeline import make_pipeline
# from sklearn.metrics import accuracy_score

# Build a pipeline with TruncatedSVD and SVM classifier
# Specify the best n_components hyperparameter.
# model = make_pipeline(TruncatedSVD(n_components=75), SVC())
max=0
maxIndex=0

for i in range(1,97):
  model = make_pipeline(TruncatedSVD(n_components=i), SVC())

  # Fit the model on training data
  model.fit(X_train_embeddings, y_train)

  # Make predictions on test data
  predictions = model.predict(X_test_embeddings)

  # Evaluate the accuracy
  accuracy = accuracy_score(y_test, predictions)
  if accuracy > max:
    max = accuracy
    maxIndex=i

print(f"Accuracy {maxIndex}: {max}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from nltk.corpus import movie_reviews
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
# Load BERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
bert_model = AutoModel.from_pretrained("bert-base-uncased")
```

```python
def calculate_embeddings(texts):
    embeddings = []
    for text in tqdm(texts):
        # Tokenize the text
        tokens = tokenizer(text, return_tensors="pt", truncation=True, padding=True)

        # Get BERT model output
        with torch.no_grad():
            model_output = bert_model(**tokens)

        # ///////

        # Extract the token embeddings
        token_embeddings = model_output.last_hidden_state

        # Compute the mask to ignore padded elements in the embeddings.
        input_mask_expanded = tokens['attention_mask'].unsqueeze(-1).expand(token_embeddings.size()).float()

        # Sum embeddings across the sequence length
        sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, 1)

        # Count non-padding tokens for each sequence in the batch
        sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)

        # Mean pooling: sum of embeddings divided by the number of non-padding tokens
        doc_vector = sum_embeddings / sum_mask


        # Use mean pooling to get sentence embeddings (is in next part)
        # ///////
        # doc_vector = # Please complete this
        embeddings.append(doc_vector)

    return np.array(embeddings)


# Calculate embeddings for training and testing sets
X_train_embeddings = calculate_embeddings(X_train)
X_test_embeddings = calculate_embeddings(X_test)
```

```python
# Build a pipeline with TruncatedSVD and SVM classifier
# Specify the best n_components hyperparameter.

# converting 3D embeddings to 2D by mean pooling
if len(X_train_embeddings.shape) == 3:
    # Mean pooling across the sequence_length dimension
    X_train_embeddings = X_train_embeddings.mean(axis=1)
if len(X_test_embeddings.shape) == 3:
    X_test_embeddings = X_test_embeddings.mean(axis=1)

max=0
maxIndex=0

for i in range(1,97):
  model = make_pipeline(TruncatedSVD(n_components=i), SVC())

  # Fit the model on training data
  model.fit(X_train_embeddings, y_train)

  # Make predictions on test data
  predictions = model.predict(X_test_embeddings)

  # Evaluate the accuracy
  accuracy = accuracy_score(y_test, predictions)
  if accuracy > max:
    max = accuracy
    maxIndex=i
  # print(f"Accuracy {i}: {accuracy}")

print(f"Maximum Accuracy {maxIndex}: {max}")
```
