```python
from evaluate import load
from hazm import word_tokenize

normalizer = Normalizer()

# for using wer we need to ensure that the number of predictions and references are equal
# Here we are taking the minimum length to ensure both lists are of the same length and split the strings into lists of words
Harf_Output_Normalized_words = word_tokenize(normalizer.normalize(harfOutput))
Original_Subtitle_Normalized_words = word_tokenize(normalizer.normalize(turingSubtitle))

equal_len=min(len(Harf_Output_Normalized_words), len(Original_Subtitle_Normalized_words))

wer = load("wer")
wer_score = wer.compute(predictions=Harf_Output_Normalized_words[:equal_len],references=Original_Subtitle_Normalized_words[:equal_len])
```

```python
print(wer_score)
```

```python
cer = load("cer")
cer_score = cer.compute(predictions=Harf_Output_Normalized_words[:equal_len],references=Original_Subtitle_Normalized_words[:equal_len])
```

```python
print(cer_score)
```

Calculating verbs and adverbs count inside the input texts:


```python
from hazm import *

# POS tagging
tagger = POSTagger(model='/content/drive/MyDrive/Uni/NLP/hasm_workshop/pos_tagger.model')
Harf_Output_tags = tagger.tag(Harf_Output_Normalized_words)

# Print tagged tokens to verify
print(Harf_Output_tags)
```

```python
# Initialize counters
verb_count = 0
adverb_count = 0
Harf_Output_verbs=[]

# Define tags for verbs and adverbs in Persian
verb_tags = {'VERB'}  # Verbs
adverb_tags = {'ADV'}  # Adverbs

# Count verbs and adverbs
for word, tag in Harf_Output_tags:
    if tag in verb_tags:
        verb_count += 1
        Harf_Output_verbs.append(word)
    elif tag in adverb_tags:
        adverb_count += 1

print(f"Number of verbs in Harf_Output_tags: {verb_count}")
print(f"Number of adverbs Harf_Output_tags: {adverb_count}")
```

lets do it all again for the original subtitle:

```python
Original_Subtitle_tags = tagger.tag(Original_Subtitle_Normalized_words)

# Print tagged tokens to verify
print(Original_Subtitle_tags)
# Initialize counters
verb_count = 0
adverb_count = 0
Original_Subtitle_verbs =[]

# Define tags for verbs and adverbs in Persian
verb_tags = {'VERB'}  # Verbs
adverb_tags = {'ADV'}  # Adverbs

# Count verbs and adverbs
for word, tag in Original_Subtitle_tags:
    if tag in verb_tags:
        verb_count += 1
        Original_Subtitle_verbs.append(word)
    elif tag in adverb_tags:
        adverb_count += 1

print(f"Number of verbs in Original_Subtitle_tags: {verb_count}")
print(f"Number of adverbs Original_Subtitle_tags: {adverb_count}")
```
