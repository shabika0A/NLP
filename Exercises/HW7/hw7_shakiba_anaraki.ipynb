```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import bitsandbytes as bnb

device = torch.device('cuda')


# Load the model and tokenizer
model_id1 = "NousResearch/Meta-Llama-3-8B-Instruct"
tokenizer1 = AutoTokenizer.from_pretrained(model_id1)

model1 = AutoModelForCausalLM.from_pretrained (
      model_id1,
      quantization_config = q_config,
      device_map="cuda",
      torch_dtype="auto",
      trust_remote_code = True
  )
```

```python
def get_context(examples, n_shots):
    messages = [
        {"role": "system", "content": "آیا این دو سوال به هم شباهت دارند؟ با بله یا خیر جواب بده."}
    ]
    for i in range(n_shots):
        example = examples[i]
        user_message = {"role": "user", "content": f"اول: {example['question1']} دوم: {example['question2']}"}
        assistant_message = {"role": "assistant", "content": f"MetaLlama: {example['label']}"}
        messages.append(user_message)
        messages.append(assistant_message)
    return messages
```

```python
CONTEXT_ZERO_SHOT1 = ""
CONTEXT_ONE_SHOT1 = get_context(examples1, 1)
CONTEXT_FIVE_SHOT1 = get_context(examples1, 5)
```

```python
def run_inference1(model, tokenizer, messages, prompt):
    context = ""
    for message in messages:
        role = "You" if message["role"] == "user" else "MetaLlama"
        context += f"{role}: {message['content']}\n"
    context += f"You: {prompt}\nMetaLlama: "

    input_tokens = tokenizer(context, return_tensors="pt")
    input_tokens = input_tokens.to(device)
    generate_ids = model.generate(**input_tokens, max_new_tokens=512, do_sample=False, repetition_penalty=1.1)
    model_output = tokenizer.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]
    return model_output[len(context):]
```

```python
# Prompt for the questions
PROMPT1 = "آیا این دو سوال به هم شباهت دارند؟"

# Evaluate the model with zero-shot, one-shot, and five-shot contexts
output_zero_shot1 = run_inference1(model1, tokenizer1, CONTEXT_ZERO_SHOT1, PROMPT1)
print(f"Zero-shot output:\n{output_zero_shot1}")

output_one_shot1 = run_inference1(model1, tokenizer1, CONTEXT_ONE_SHOT1, PROMPT1)
print(f"One-shot output:\n{output_one_shot1}")

output_five_shot1 = run_inference1(model1, tokenizer1, CONTEXT_FIVE_SHOT1, PROMPT1)
print(f"Five-shot output:\n{output_five_shot1}")
```

```python
def zero_shot_evaluation1(model, tokenizer, pairs):

    positive_results={"same","yes","یک معنی", "بله"}
    negative_results={"different","no","متفاوت", "خیر", "نه"}
    results = []

    for question1, question2 in pairs:
        prompt = f"اول: {question1} دوم: {question2}"
        result = run_inference1(model, tokenizer, [], prompt)
        print(f"prompt:{prompt}")
        print(f"result:{result}")

        if any(pos in result for pos in positive_results):
            results.append("بله")
        elif any(neg in result for neg in negative_results):
            results.append("خیر")
        else:
            results.append(result)

    return results
```

```python
def few_shot_evaluation1(model, tokenizer, pairs, n_shots, data):

    positive_results={"same","yes","یک معنی", "بله"}
    negative_results={"different","no","متفاوت", "خیر", "نه"}
    results = []

    # Use the initial examples to create the context
    context = get_context(data, n_shots)

    for question1, question2 in pairs:
        prompt = f"اول: {question1} دوم: {question2}"
        result = run_inference1(model, tokenizer, context, prompt)

        if any(pos in result for pos in positive_results):
            results.append("بله")
        elif any(neg in result for neg in negative_results):
            results.append("خیر")
        else:
            results.append(result)

    return results
```

```python
import json

# Download the JSON Lines file
!wget https://raw.githubusercontent.com/persiannlp/parsinlu/master/data/qqp/train.jsonl -O train.jsonl

# Function to read the file and split into smaller parts
# making 10 sets for data
items_per_set = 10
whole_data = []
with open('train.jsonl', 'r', encoding='utf-8') as file:
    for line in file:
        whole_data.append(json.loads(line))

# Split data into sets
sets = []
for i in range(10):
    start_index = i * items_per_set
    end_index = start_index + items_per_set
    sets.append(whole_data[start_index:end_index])

data = sets[2]

# print(data)
for d in data :
  print(d)
  print()

pairs = [(entry['q1'], entry['q2']) for entry in data]
labels = [entry['label'] for entry in data]
print(labels)
```

```python
# import torch
# from transformers import AutoTokenizer, AutoModelForCausalLM

# device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Function to run inference
def run_inference(model, tokenizer, context, question1, question2):
    TEMPLATE = "{context}You: آیا این دو سوال از نظر معنا مشابه هستند؟ با بله یا خیر جواب بده. اول: {question1} دوم: {question2}\nPersianMind: "
    model_input = TEMPLATE.format(context=context, question1=question1, question2=question2)
    inputs = tokenizer(model_input, return_tensors="pt", padding=True, truncation=True)
    inputs = inputs.to(device)

    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=50, do_sample=False, repetition_penalty=1.1)
        generated_text = tokenizer.batch_decode(outputs, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]
        generated_response = generated_text[len(model_input):].strip()
    return generated_response



# Zero-shot evaluation function
def zero_shot_evaluation(model, tokenizer, pairs):
    positive_results={"یک معنی","آره","بله"}
    negative_results={"متفاوت","خیر","نه"}
    results = []
    for question1, question2 in pairs:
        result = run_inference(model, tokenizer, "", question1, question2)
        # Check if "بله" or "خیر" is inside the result
        #if "بله" in result:
        if any(pos in result for pos in positive_results):
            results.append("بله")
        elif any(pos in result for pos in negative_results):
        #if "خیر" in result:
            results.append("خیر")
        else:
            results.append(result)
    return results

# Few-shot evaluation function
def few_shot_evaluation(model, tokenizer, pairs, n_shots):
    positive_results={"یک معنی","بله"}
    negative_results={"متفاوت","خیر","نه"}

    results = []

    def get_context(examples, n_shots):
        context = ""
        for i in range(n_shots):
            example = examples[i]
            context += f"آیا این دو سوال از نظر معنا مشابه هستند؟ با بله یا خیر جواب بده. اول: {example['q1']} دوم: {example['q2']}\nPersianMind: {'بله' if example['label'] == '1' else 'خیر'}\n"
        return context

    # Use the initial examples to create the context
    context = get_context(data, n_shots)

    for question1, question2 in pairs:
        result = run_inference(model, tokenizer, context, question1, question2)
        # Check if "بله" or "خیر" is inside the result
        #if "بله" in result:
        if any(pos in result for pos in positive_results):
            results.append("بله")
        elif any(pos in result for pos in negative_results):
        #if "خیر" in result:
            results.append("خیر")
        else:
            results.append(result)
    return results


# Zero-shot evaluation
results_zero_shot = zero_shot_evaluation(model, tokenizer, pairs)
print("Zero-shot Results:", results_zero_shot)

# One-shot evaluation
results_one_shot = few_shot_evaluation(model, tokenizer, pairs, n_shots=1)
print("One-shot Results:", results_one_shot)

# Five-shot evaluation
results_five_shot = few_shot_evaluation(model, tokenizer, pairs, n_shots=5)
print("Five-shot Results:", results_five_shot)

# Compare results with labels
def compare_results(results, labels):
    correct = 0
    for result, label in zip(results, labels):
        expected = "بله" if label == '1' else "خیر"
        if expected in result:
            correct += 1
    accuracy = correct / len(labels)
    return accuracy

accuracy_zero_shot = compare_results(results_zero_shot, labels)
accuracy_one_shot = compare_results(results_one_shot, labels)
accuracy_five_shot = compare_results(results_five_shot, labels)

print(f"Zero-shot Accuracy: {accuracy_zero_shot * 100:.2f}%")
print(f"One-shot Accuracy: {accuracy_one_shot * 100:.2f}%")
print(f"Five-shot Accuracy: {accuracy_five_shot * 100:.2f}%")
```
